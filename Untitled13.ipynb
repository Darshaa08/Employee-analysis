```python
print(df["LeaveOrNot"].value_counts())


# 0- left the company and 1 - Stayed in the company
```

```python
df["LeaveOrNot"].value_counts().plot(kind="bar")
plt.title("Employee Attrition")
plt.xlabel("LeaveOrNot")
plt.ylabel("Count")
plt.show()
```

Employees Attribution Analysis -
Most of the Employees stayed in the company only Employee retention is higher than attrition, with 1,676 employees staying compared to 1,088 leaving.
Despite the higher retention rate, employee turnover remains notable and warrants further investigation.

```python
print(df["Gender"].value_counts())

df["Gender"].value_counts().plot(kind="bar")
plt.title("Gender Distribution")
plt.show()
```

Gender Distribution -
Male employees slightly outnumber female employees, representing the larger share of the workforce.

```python
print(df["City"].value_counts())

df["City"].value_counts().plot(kind="bar")
plt.title("Employees by City")
plt.show()
```

Employees by City-
Bangalore employs the largest number of employees, indicating it is the organization's major operational location.

```python
print(df["Education"].value_counts())
print(df["PaymentTier"].value_counts())
```

Education Based on the Degree-
Bachelor's degree holders form the majority of the workforce.

```python
print(df.groupby("Gender")["LeaveOrNot"].mean())
```

Attrition Rate by Gender-
Attrition is significantly higher among female employees than male employees.

```python
print(df.groupby("City")["LeaveOrNot"].mean())
```

Attrition Rate by City-
Employees in Pune exhibit the highest attrition rate among all cities.

```python
print(df.info())
print(df.shape)
print(df.isnull().sum())
```

```python
print(df.groupby("Education")["LeaveOrNot"].mean())
```

```python
print(df.groupby("PaymentTier")["LeaveOrNot"].mean())

df.groupby("PaymentTier")["LeaveOrNot"].mean().plot(kind="bar")
plt.title("Attrition by Payment Tier")
plt.show()
```

Attrition Rate by Payment Tier-
Payment Tier 2 employees exhibit the highest attrition among all salary tiers.

```python
print(df.groupby("ExperienceInCurrentDomain")["LeaveOrNot"].mean())

df.groupby("ExperienceInCurrentDomain")["LeaveOrNot"].mean().plot()
plt.title("Attrition by Experience")
plt.show()
```

```python
print(df.groupby("EverBenched")["LeaveOrNot"].mean())

df.groupby("EverBenched")["LeaveOrNot"].mean().plot(kind="bar")
plt.title("Attrition by Bench Status")
plt.show()
```

Attrition Rate by Bench Status-

Employees who were benched are more likely to leave the organization than those who were not benched.

```python
df["Age"].hist()
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

Age Distribution

The majority of employees are between 25 and 30 years old, indicating a relatively young workforce.

Conclusion:

The analysis revealed that employee attrition is influenced by factors such as payment tier, bench status, experience level, and location.
